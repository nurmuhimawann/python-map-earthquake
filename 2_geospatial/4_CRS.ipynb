# UTM => Latitude Longitude

```python
import utm

easting = 166072
northing = 110682
zone_number = 50  # You need to specify the zone number
zone_letter = 'N'  # And the zone letter

lat, lon = utm.to_latlon(easting, northing, zone_number, zone_letter)

print(f"Latitude: {lat}, Longitude: {lon}")
```

# Latitude Longitude => UTM

```python
latitude = 1
longitude = 114

# Convert to UTM
easting, northing, zone_number, zone_letter = utm.from_latlon(latitude, longitude)

print(f"Easting: {easting}, Northing: {northing}, Zone Number: {zone_number}, Zone Letter: {zone_letter}")
```

# Using Geopandas

Jika kita ingin mengkonversi koordinat sistem dengan geopandas kita harus mendefinisikan terlebih dahulu koordinat sistem data awalnya, dan mendefinisikan koordinat sistem targetnya.

<h3>NOTE:</h3>

**Dalam satu geodataframe hanya akan memiliki satu CRS. Apabila data memiliki CRS yang berbeda-beda maka harus di convert terlebih dahulu sebelum digabungkan menjadi satu geodataframe**

```python
import geopandas as gpd
from shapely.geometry import Polygon

utm_polygons = [
    Polygon([(500000, 4649776), (500500, 4649776), (500500, 4649276), (500000, 4649276), (500000, 4649776)])
]

# Define the UTM CRS
zone_number = 32
zone_letter = 'N'
utm_crs = f"+proj=utm +zone={zone_number} +datum=WGS84 +units=m +{'north' if zone_letter.upper() >= 'N' else 'south'}"
gdf_utm = gpd.GeoDataFrame(geometry=utm_polygons, crs=utm_crs)

# Convert to latitude and longitude
gdf_latlon = gdf_utm.to_crs("EPSG:4326")

gdf_latlon.explore()
```
